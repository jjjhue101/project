```python
import plotly.figure_factory as ff
from matplotlib import pyplot as plt
import plotly.express as py
import pandas as pd
import plotly.tools as tls
import plotly.graph_objects as go
```

Black Carbon

```python
scope = ['Poland']
df=pd.read_csv('../Downloads/covid_p/bc_df.csv',usecols= \
               ['timestamp','city', 'country', 'value','latitude', 'longitude'])
df.head(10)
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=df['city'],
                y=df['value'],
            
                marker_color='green'
                ))
fig.update_layout(
    xaxis_tickfont_size=14,
    yaxis=dict(
        titlefont_size=16,
        tickfont_size=14,
        range=[0,2],
    ),
    legend=dict(
        x=0,
        y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgb(255, 255, 255, 0)'
    ),
    barmode='stack',
    bargap=0.15, # gap between bars of adjacent location coordinates.
    bargroupgap=0.1 # gap between bars of the same location coordinate.
)
fig.show()
```

```python
data=[dict(type='choropleth',autocolorscale=False,locations=df['city'],z=df['value'],locationmode='USA-states',text=df['value'],colorscale='greens',colorbar=dict(title="black Carbon value"))]
layout=dict(title="AQI-BC",geo=dict(scope='world',projection=dict(type='miller'),showlakes=True,lakecolor='rgb(66,245,66)'))
```

```python
fig=dict(data=data,layout=layout)
import plotly.io as pio

pio.show(fig)
```
